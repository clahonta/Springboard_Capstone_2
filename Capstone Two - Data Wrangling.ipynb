```python
data = data.reset_index()
```

```python
globe_survey_dict = {1:"Strongly agree",2:"Agree somewhat",3:"Agree slightly",4:"Neither agree nor disagree",5:"Disagree slightly",6:"Disagree somewhat",7:"Disagree strongly"}
```

```python
globe_survey_cols = list(data.columns[24:])
```

```python
data[globe_survey_cols] = data[globe_survey_cols].replace(globe_survey_dict)
```

```python
mrq_dict = {1:"Yes",2:"Rather yes",3:"Neither yes or no",4:"Rather no",5:"No"}
```

```python
mrq_cols = list(data.columns[12:21])
```

```python
data[mrq_cols] = data[mrq_cols].replace(mrq_dict)
```

```python
data['sex'] = data['sex'].replace({1:"Male",2:"Female"})
```

```python
data.head()
```

```python
data['religion'] = data['religion'].replace({1:"Protestant",2:"Catholic",3:"Jewish",4:"Muslim",5:"Buddhist",6:"None",7:"Jehovah",8:"Evangelic",9:"Spiritualism",10:"Other",11:"Orthodox",12:"Hindu"})
```

```python
data['edu_level'] = data['edu_level'].replace({1:"No formal education",2:"Primary school",3:"Secondary school",4:"High school or technical school",5:"Bachelors or Masters degree"})
```

```python
data['pension'] = data['pension'].replace(globe_survey_dict)
```

```python
data['material_situation'] = data['material_situation'].replace({1:"Much better than average",2:"Better than average",3:"Average",4:"Worse than average",5:"Much worse than average"})
```
